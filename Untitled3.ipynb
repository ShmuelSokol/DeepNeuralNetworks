```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
directory = 'C:/Users/shmue/'
```

```python
df = pd.read_csv(directory+'housepricedata.csv')
```

```python
df
```

```python
dataset = df.values
```

```python
dataset
```

```python
X = dataset[:,0:10]
```

```python
Y = dataset[:,10]
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
```

```python
X_scale
```

```python
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, Y, test_size=0.3)
```

```python
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5)
```

```python
print(X_train.shape, X_val.shape, X_test.shape, Y_train.shape, Y_val.shape, Y_test.shape)
```

```python
model = Sequential([    Dense(32, activation='relu', input_shape=(10,)),    Dense(32, activation='relu'),    Dense(1, activation='sigmoid'),])
```

```python
model.compile(optimizer='sgd',              loss='binary_crossentropy',              metrics=['accuracy'])
```

```python
hist = model.fit(X_train, Y_train,          batch_size=32, epochs=50,          validation_data=(X_val, Y_val))
```

```python
model.evaluate(X_test, Y_test)[1]
```

```python
results = model.predict(X_scale)
```

```python
results = results.tolist()
for n in range(len(results)):
    results[n] = results[n][0]
df['prediction'] = results
```

```python
df
```

```python
df.to_csv(directory+'predictions.csv')
```
